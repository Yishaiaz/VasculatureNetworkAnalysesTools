```python
import os

```

```python
path_to_properties_exploration_dir = 'properties_exploration/'
assert os.path.isdir(path_to_properties_exploration_dir)
path_to_vertices_exploration_dir = os.path.join(path_to_properties_exploration_dir, 'vertices_properties_exploration')
assert os.path.isdir(path_to_vertices_exploration_dir)
path_to_edges_exploration_dir = os.path.join(path_to_properties_exploration_dir, 'edges_properties_exploration')
assert os.path.isdir(path_to_edges_exploration_dir)
```

```python
import matplotlib.pyplot as plt
import pickle

def single_prop_exploration(exploration_type_dir: str, prop_f_name: str):
    path_to_edge_property = os.path.join(exploration_type_dir, prop_f_name)
    with open(path_to_edge_property, 'rb') as f:
        prop_array = pickle.load(f)
    print(path_to_edge_property)
    print(prop_array.max(), prop_array.min(), prop_array.mean(), prop_array.std())
    fig, ax = plt.subplots()
    ax.hist(prop_array)
    ax.set_title(prop_f_name)
    
for prop_fname in ('radii_all_vals', 'length_all_vals' ,'artery_binary_all_vals'):
    single_prop_exploration(path_to_edges_exploration_dir, prop_fname)

    print('#' * 30)

for prop_fname in ('annotation_all_vals', 'artery_binary_all_vals' ,'distance_to_surface_all_vals', 'radii_all_vals'):
    single_prop_exploration(path_to_vertices_exploration_dir, prop_fname)
```
